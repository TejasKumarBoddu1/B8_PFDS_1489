# Quetion 1

```python
import numpy as np
```

```python
marks = np.array([
    [85, 90, 78],
    [92, 88, 95],
    [76, 80, 72],
    [89, 94, 91],
    [65, 70, 68]
])
```

```python
print("Student Marks (2D Array):\n", marks)
```

```python
avg_marks = np.mean(marks, axis=1)
min_marks = np.min(marks, axis=1)
max_marks = np.max(marks, axis=1)
```

```python

print("\nAverage Marks per Student:", avg_marks)
print("Minimum Marks per Student:", min_marks)
print("Maximum Marks per Student:", max_marks)
```

```python

print("\nMarks of 1st Student:", marks[0])
print("Marks of 3rd Student:", marks[2])
```

```python

above_80 = marks > 80
print("\nBoolean Mask (Scores > 80):\n", above_80)
print("\nStudents with Scores > 80:\n", marks[above_80])
```

```python
reshaped = marks.T
print("Subject-wise Performance \n", reshaped)
```

# Quetion 2

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Diya", "Karan", "Meera", "Rohit"],
    "ID": [101, 102, 103, 104, 105],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"],
    "Salary": [45000, 60000, 52000, 75000, 48000]
}

df = pd.DataFrame(data)
df
```

```python
high_salary = df[df["Salary"] > 50000]
print("Employees with Salary > 50,000:\n")
high_salary
```

```python
sorted_df = df.sort_values(by="Salary", ascending=False)
print("DataFrame Sorted by Salary (Descending):\n")
sorted_df
```

```python
df["Bonus"] = df["Salary"] * 0.10
print("DataFrame with Bonus Column:\n")
df
```

```python
df["Total"] = df["Salary"] + df["Bonus"]
total_expense = df["Total"].sum()
print("Total Salary Expense (Including Bonuses):", total_expense)
```
